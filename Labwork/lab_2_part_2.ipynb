### Task 1
Для трьох введених натуральних чисел (користувач вводить числа через пробіл) знайдіть мінімальне та максимальне

````python
def min_max_number():
    while True:
        try:
            user_input = input('Введіть 3 числа через пробіл:\n')
            numbers = [int(x) for x in user_input.split()]
            break
        except ValueError:
            print('Потрібно вводити число. Спробуйте ще раз.')
    result = print(f'min is {min(numbers)}\nmax is {max(numbers)}')
    return result

min_max_number()
````

### Task 2
* Програма приймає на вхід два параметри: стать спортсмена(`sex = (f, m)`) та вік спортсмена(`age = [0,100]`).  
* Спортсмен-дівчина може брати участь у змаганні, якщо її вік більший 18, спортсмен-хлопець - якщо вік більший 16
* Якщо введеніть не коректні дані, виведіть повідомлення про помилку

````python
class SexError(Exception):
    pass

def sex():
    '''Determines sex of participant'''
    while True:
        try:
            sex = input("Вкажіть вашу стать:\n'm', якщо чоловік, 'f', якщо жінка\n")
            for gender in sex:
                if gender == 'f' or gender == 'm':
                    return sex
                else:
                    raise SexError
            break
        except SexError:
            print(f"Помилка! Оберіть 'm' або 'f'\n")
    return sex
   
def age():
    '''Determines age of participant'''    
    while True:
        try:
            age = int(input(f'Вкажіть скільки вам повних років:\n'))
            break
        except ValueError:
            print('Потрібно вводити число. Спробуйте ще раз.')
    return age

def competition(sex, age):
    '''Decides whether a person can take part in the competition'''
    if sex == 'f' and age < 18:
        age_y = int(18 - age)
        result = f'Нажаль ви не можете приймати участь у змаганні, будемо раді вас бачити за {age_y} роки'
    elif sex == 'm' and age < 16:
        age_y = int(16 - age) 
        result = f'Нажаль ви не можете приймати участь у змаганні, будемо раді вас бачити за {age_y} роки'
    elif age > 100:
        result = 'Вік учасника має бути менше 100 років. Нажаль ви не можете приймати участь у змаганні'
    else:
        result = 'Вітаємо! Ви зареєстровані для участі у змаганні'
    return result

sex = sex()
age = age()
competition(sex, age)
````

### Task 3
Для введеного натурального числа побудуйте драбинку.  
> _Наприклад, якщо N = 5, драбинка буде мати вигляд:_  
    &  
    &&  
    &&&  
    &&&&  
    &&&&&

````python
def ladder():
    '''Сonverts a number to a ladder'''
    number = int(input('Введіть число і я перетворю його на драбинку!: '))
    ladder = ''
    for x in range(1, number + 1):
        ladder += '&'
        print(ladder)    
    return()

ladder()
````

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

_Послідовність чисел генеруємо рандомно_
```python
import random
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]
```

````python
import random
import numpy as np
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

def min_max(list_with_random_numbers):
    min_num = min(list_with_random_numbers)
    max_num = max(list_with_random_numbers)
    print(f'minimum number - {min_num}\nmaximum number - {max_num}')
    return min_num, max_num

def average_value(list_with_random_numbers):
    average = round(sum(list_with_random_numbers)/len(list_with_random_numbers), 2)
    print(f'Average value of the sequence is equal to {average}')
    return average

def median_seq(list_with_random_numbers):
    sorted_list = sorted(list_with_random_numbers)
    median = (sorted_list[499] + sorted_list[500]) / 2
    print(f'Median = {median}')
    return median

def frequency_num(list_with_random_numbers):
    frequency_dict = {}
    for number in list_with_random_numbers:
        frequency_dict[number] = list_with_random_numbers.count(number)
    for number, value in sorted(frequency_dict.items()):
        print(f"{number}: {value}")
    return frequency_dict

def standard_deviation(list_with_random_numbers):
    deviation = round(np.std(list_with_random_numbers), 2)
    print(f'Standard deviation = {deviation}')
    return deviation

min_max(list_with_random_numbers)
average_value(list_with_random_numbers)
median_seq(list_with_random_numbers)
frequency_num(list_with_random_numbers)
standard_deviation(list_with_random_numbers)
````
